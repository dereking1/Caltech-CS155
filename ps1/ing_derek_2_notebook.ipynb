## Example code using the polyfit and kfold functions

#### Note: This section is not part of the homework problem, but provides some potentially-helpful example code regarding the usage of numpy.polyfit, numpy.polyval, and sklearn.model_selection.KFold.

First, let's generate some synthetic data: a quadratic function plus some Gaussian noise.

```python
# Coefficients of the quadratic function, y(x) = ax^2 + bx + c:
a = 2
b = 5
c = 7

N = 200       # Number of data points
x = np.linspace(-10, 10, num = N)                # x ranges from -10 to 10
# y is the quadratic function of x specified by a, b, and c, plus noise
y = a*x**2 + b*x + c + 15* np.random.randn(N)

# Plot the data:
plt.figure()
plt.plot(x, y, marker = 'o', linewidth = 0)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic data')

plt.show()
```

Next, we'll use the numpy.polyfit function to fit a quadratic polynomial to this data. We can evaluate the resulting polynomial at arbitrary points.

```python
# Fit a degree-2 polynomial to the data:
degree = 2
coefficients = np.polyfit(x, y, degree)

# Print out the resulting quadratic function:
print('We fit the following quadratic function: f(x) = %f*x^2 + %f*x + %f' %  \
      (coefficients[0], coefficients[1], coefficients[2]))

# Evaluate the fitted polynomial at x = 4:
x_test = 4
f_eval = np.polyval(coefficients, x_test)
print('\nf(%i) = %f' % (x_test, f_eval))

# Let's visualize our fitted quadratic:
plt.figure()

plt.plot(x, y, marker = 'o', linewidth = 0)
plt.plot(x, np.polyval(coefficients, x), color = 'red', linewidth = 3)

plt.legend(['Data', 'Fitted quadratic'], loc = 'best')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic data with Fitted Quadratic')

plt.show()
```

## Loading the Data for Problem 2

This code loads the data from bv_data.csv using the load_data helper function. Note that data[:, 0] is an array of all the x values in the data and data[:, 1] is an array of the corresponding y values.

```python
def load_data(filename):
    """
    Function loads data stored in the file filename and returns it as a numpy ndarray.
    Input:
        filename: given as a string.
    Output:
        Data contained in the file, returned as a numpy ndarray
    """
    return np.loadtxt(filename, skiprows=1, delimiter=',')
```

```python
data = load_data('https://raw.githubusercontent.com/charlesincharge/Caltech-CS155-2022/main/sets/set1/data/bv_data.csv')
x = data[:, 0]
y = data[:, 1]
```

Write your code below for solving problem 2 part B:

```python
degrees = [1, 2, 6, 12]
N_vals = np.linspace(20,100,17,dtype=int)
n_folds = 5
kfold = KFold(n_splits=n_folds)

fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))

for d in degrees:
    train_err_d = []
    valid_err_d = []
    for N in N_vals:
        x_N = x[:N]
        y_N = y[:N]
        train_err_N = 0
        valid_err_N = 0
        for train_idx, valid_idx in kfold.split(x_N):
            x_train, x_valid = x_N[train_idx], x_N[valid_idx]
            y_train, y_valid = y_N[train_idx], y_N[valid_idx]

            p = np.polyfit(x_train, y_train, d)
            train_eval = np.polyval(p, x_train)
            valid_eval = np.polyval(p, x_valid)

            train_err = np.mean(np.square(np.subtract(train_eval, y_train)))
            valid_err = np.mean(np.square(np.subtract(valid_eval, y_valid)))
            train_err_N += train_err
            valid_err_N += valid_err

        train_err_d.append(train_err_N / n_folds)
        valid_err_d.append(valid_err_N / n_folds)
    if d == 1:
        ax1.plot(N_vals, train_err_d, label='Training error')
        ax1.plot(N_vals, valid_err_d, label='Validation error')
        ax1.set_xlabel('N')
        ax1.set_ylabel('Average Error')
        ax1.set_ylim(bottom=0, top=5)
        ax1.set_title('Degree = 1')
        ax1.legend(loc="upper right")
    elif d == 2:
        ax2.plot(N_vals, train_err_d, label='Training error')
        ax2.plot(N_vals, valid_err_d, label='Validation error')
        ax2.set_xlabel('N')
        ax2.set_ylabel('Average Error')
        ax2.set_ylim(bottom=0, top=5)
        ax2.set_title('Degree = 2')
        ax2.legend(loc="upper right")
    elif d == 6:
        ax3.plot(N_vals, train_err_d, label='Training error')
        ax3.plot(N_vals, valid_err_d, label='Validation error')
        ax3.set_xlabel('N')
        ax3.set_ylabel('Average Error')
        ax3.set_ylim(bottom=0, top=5)
        ax3.set_title('Degree = 6')
        ax3.legend(loc="upper right")
    elif d == 12:
        ax4.plot(N_vals, train_err_d, label='Training error')
        ax4.plot(N_vals, valid_err_d, label='Validation error')
        ax4.set_xlabel('N')
        ax4.set_ylabel('Average Error')
        ax4.set_ylim(bottom=0, top=5)
        ax4.set_title('Degree = 12')
        ax4.legend(loc="upper right")

```
