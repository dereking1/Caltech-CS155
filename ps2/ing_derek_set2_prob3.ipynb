```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
train_file = 'https://raw.githubusercontent.com/charlesincharge/Caltech-CS155-2022/main/sets/set2/data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

## Lasso Regression

```python
alphas = np.linspace(0.01, 3, 30)
W_lasso = np.zeros((30, 9))
for i,a in enumerate(alphas):
    lasso = Lasso(alpha=a)
    lasso.fit(X_train, y_train)
    w = lasso.coef_
    W_lasso[i] = w

plt.figure()
for j in range(9):
    plt.plot(alphas, W_lasso[:,j], label=f'w{j+1}')

plt.xlabel('Alpha')
plt.ylabel('Weight')
plt.title('Model Weights with Lasso Regression for Different Values of Alpha')
plt.legend(loc='upper right')
plt.show()
```

## Ridge Regression

```python
alphas = np.arange(1, 1e4 + 1)
W_ridge = np.zeros((len(alphas), 9))
for i,a in enumerate(alphas):
    ridge = Ridge(alpha=a)
    ridge.fit(X_train, y_train)
    w = ridge.coef_
    W_ridge[i] = w

plt.figure()
for j in range(9):
    plt.plot(alphas, W_ridge[:,j], label=f'w{j+1}')

plt.xlabel('Alpha')
plt.ylabel('Weight')
plt.title('Model Weights with Ridge Regression for Different Values of Alpha')
plt.legend(loc='upper right')
plt.show()
```
